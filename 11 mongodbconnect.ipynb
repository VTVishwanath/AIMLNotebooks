```python
connection = MongoClient("mongodb+srv://testuser:4nG8IsQgTVk4C4Yu@cluster0.2xbo6fr.mongodb.net/")
```

```python
client = connection
database = client["sample_analytics"]
collection = database["transactions"]

# Created with Studio 3T, the IDE for MongoDB - https://studio3t.com/

pipeline = [
    {
        u"$match": {
            u"account_id": {
                u"$gt": Int64(500000)
            }
        }
    }, 
    {
        u"$group": {
            u"_id": {
                u"account_id": u"$account_id"
            },
            u"SUM(transaction_count)": {
                u"$sum": u"$transaction_count"
            }
        }
    }, 
    {
        u"$project": {
            u"account_id": u"$_id.account_id",
            u"SUM(transaction_count)": u"$SUM(transaction_count)",
            u"_id": 0
        }
    }, 
    {
        u"$sort": SON([ (u"account_id", 1) ])
    }
]

cursor = collection.aggregate(pipeline,allowDiskUse = True)
try:
    for doc in cursor:
        print(doc)
finally:
    client.close()
```
