```python
# Supress Warnings
import warnings
warnings.filterwarnings('ignore')

# Import the numpy and pandas package
import numpy as np
import pandas as pd

# Data Visualisation
import matplotlib.pyplot as plt 
import seaborn as sns
import os
```

```python
housing = pd.DataFrame(pd.read_csv("Housing.csv"))
```

```python
# Check the head of the dataset
housing.head()
```

```python
housing.columns
```

```python
grouped = housing[['price', 'area', 'bedrooms', 'bathrooms', 'stories','furnishingstatus']].groupby('furnishingstatus')
```

```python
grouped
```

```python
mean_price = grouped.mean().reset_index()
```

```python
mean_price.head()
```

```python
# Create the pie chart
mean_price.plot(kind='pie', y='price', labels=mean_price['furnishingstatus'], startangle=90) #autopct='%1.1f%%'

# add title
plt.title('Avg price of houses based on furnishing')

# Show the chart
plt.show()
```

```python
# Create the pie chart
mean_price.plot(kind='pie', y='area', labels=mean_price['furnishingstatus'], startangle=90) #autopct='%1.1f%%'

# add title
plt.title('Avg area of houses based on furnishing')

# Show the chart
plt.show()
```

```python
grouped = housing.groupby('furnishingstatus')
counted = grouped.count().reset_index()
counted.head()
```

```python
# Create the pie chart
counted.plot(kind='pie', y='area', labels=counted['furnishingstatus'], autopct='%1.1f%%', startangle=45)

# add title
plt.title('No of houses in % based on furnishing')

# Show the chart
plt.show()
```

```python
# create donut chart
fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))

wedges, texts, autotexts = ax.pie(counted['area'], wedgeprops=dict(width=0.6), startangle=60,
                                  autopct='%1.1f%%', textprops=dict(color="w"))

# add labels
ax.legend(wedges, counted['furnishingstatus'],
          title="Furnishing Status",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

# add title
ax.set_title("No of houses in % based on furnishing")

# display chart
plt.show()
```

```python
# create sample data
data = {'year': [2016, 2017, 2018, 2019, 2020],
        'Toyota': [100, 120, 140, 160, 180],
        'Ford': [80, 90, 100, 110, 120],
        'Honda': [60, 70, 80, 90, 100]}
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
# create stacked column chart
plt.bar(df['year'], df['Toyota'], label='Toyota')
plt.bar(df['year'], df['Ford'], bottom=df['Toyota'], label='Ford')
plt.bar(df['year'], df['Honda'], bottom=df['Toyota']+df['Ford'], label='Honda')

# add labels and title
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Annual Sales by Car Brand')

# add legend
plt.legend()

# display chart
plt.show()
```
